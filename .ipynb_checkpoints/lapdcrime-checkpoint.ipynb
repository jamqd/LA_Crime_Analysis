## How safe is West LA?

How safe West Los Angeles is can be determined by comparing crime statistics between the 21 LAPD divisions. A relevant metric would be the number of crimes in each division which would be normalized by the population of each division to compute crimes per capita.

West Los Angeles is the 6th safest LAPD division by number of crimes per capita (0.339). This is within 0.07 crimes per capita of the safest LAPD division, Van Nuys (0.269).

Displayed below is a list and bar chart of crimes per capita for each LAPD division from safest to most dangerous.

```python
# create dictionary with crimes per capita for each LAPD division
crimes_per_capita_dict = {}
for n in num_crimes_dict:
    # divide number of crimes by population to compute crimes per capita
    crimes_per_capita_dict[n] = num_crimes_dict[n] / population_dict[n]
# create data frame for crimes per capita
cpc_df = pd.DataFrame({"name": list(crimes_per_capita_dict.keys()), "count": list(crimes_per_capita_dict.values())})
print(cpc_df.sort_values("count"))

# plot bar chart of crimes per capita for each LAPD division
plt.figure(dpi = 125)
plt.ylabel("Crimes per Capita")
plt.title("LAPD Divisions")
cpc_df.sort_values("count").plot(x="name",kind="bar",ax = plt.gca(), sort_columns=True, legend=False)
plt.xlabel("LAPD Division")
```

## Machine Learning for Predicting Crime (Experimental)

Attempting to predict the most likely crime to happen using shallow neural network. Currently, there is a bias problem as the model is not fitting to the training set well (low accuracy on the training set and test set).

```python
# create separate dataframe with data used for training
crime_train_df = crime_df[["Time Occurred", "Location ", "Crime Code Description", "Victim Age", "Victim Sex", "Victim Descent"]]
#drop nan values
crime_train_df = crime_train_df.dropna()
#shuffle dataset
crime_train_df = crime_train_df.sample(frac=1).reset_index(drop=True)
print(crime_train_df.shape)
crime_train_df.head()
```

```python
# get split for training (80%) and test (20%)
train_count = int(crime_train_df.shape[0] * 0.8)
test_count = crime_train_df.shape[0] - train_count

# generate input arrays
x_train = np.zeros((train_count, 3))
x_test = np.zeros((test_count, 3))

# add times to train and test set
x_train[:,0] += crime_train_df["Time Occurred"].values[0:train_count]
x_test[:,0] += crime_train_df["Time Occurred"].values[train_count:]

# add coordinates to train set
count = 0
for n in crime_train_df["Location "].values[0:train_count]:
    cood = make_tuple(n)
    x_train[count, 1] += cood[0]
    x_train[count, 2] += cood[1]
    count += 1

# add coordinates to test set
count = 0
for n in crime_train_df["Location "].values[train_count:]:
    cood = make_tuple(n)
    x_test[count, 1] += cood[0]
    x_test[count, 2] += cood[1]
    count += 1
    
# add victim age to training and test set
# x_train[:,3] += crime_train_df["Victim Age"].values[0:train_count]
# x_test[:,3] += crime_train_df["Victim Age"].values[train_count:]

print("x_train \n \n" + str(x_train) + "\n")
print("x_test \n \n" + str(x_test))
```

```python
# generate output arrays
y_train = np.zeros((train_count, 149))
y_test = np.zeros((test_count, 149))

# create one hot dataframe for crime codes
one_hot = pd.get_dummies(crime_train_df["Crime Code Description"])


# add one hot matrix to training and test sets
y_train += one_hot[0:train_count]
print("y_train: \n \n" + str(y_train) + "\n")
y_test += one_hot[train_count:]
print("y_test: \n \n" + str(y_test))
```

```python
# create list for crime descriptions that corresponds to one hot indexes
one_hot_list = list(one_hot)
print(one_hot_list)
```

```python
# create input layer
X_input = Input(shape=(3,),name="input")
# Dense layers
X = Dense(units=512, activation='relu',name='dense1')(X_input)
X = Dense(units=256, activation='relu',name='dense2')(X)
X = Dense(units=149, activation='softmax',name='output')(X)

# create and compile model
model = Model(inputs = X_input, outputs = X, name="crime_prediction")
model.compile('Adam', loss=categorical_crossentropy, metrics=['accuracy'])
print(model.summary())
```

```python
# train and save model
model.fit(x_train, y_train, epochs=25, validation_data = (x_test, y_test))
model.save('crime_prediction_model.h5')
```
